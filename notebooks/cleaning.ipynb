```python
import pickle
with open("../documents.pkl", "rb") as file:
    documents = pickle.load(file)
```

```python
cleaned_documents = []
```

```python
import re
def clean_text(text):
    text = text.replace("\t", " ")
    text = text.replace("\r", "\n")

    text = re.sub(r" +", " ", text)
    text = re.sub(r"\n+", "\n", text)

    return text.strip()
```

```python
for document in documents:
    cleaned_documents.append(
        {
            "source": document["source"],
            "type": document["type"],
            "content": clean_text(document["content"])
        }
    )
```
