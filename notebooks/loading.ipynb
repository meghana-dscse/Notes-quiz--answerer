```python
from pathlib import Path

from docx import Document
from pptx import Presentation
from pypdf import PdfReader
```

```python
def load_pdf(file_path):
    reader = PdfReader(file_path)

    text = ""

    for page in reader.pages:
        page_text = page.extract_text()

        if page_text:
            text += page_text + "\n"

    return text
```

```python
def load_docx(file_path):
    document = Document(file_path)

    return "\n".join(
        paragraph.text
        for paragraph in document.paragraphs
        if paragraph.text.strip()
    )
```

```python
def load_pptx(file_path):
    presentation = Presentation(file_path)

    text = []

    for slide in presentation.slides:
        for shape in slide.shapes:
            if hasattr(shape, "text"):
                if shape.text.strip():
                    text.append(shape.text)

    return "\n".join(text)
```

```python
def load_txt(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        return file.read()
```

```python
documents = []
```

```python
from tqdm import tqdm
from docx import Document

def load_docx(file_path):
    document = Document(file_path)

    text = ""

    for paragraph in tqdm(document.paragraphs, desc=file_path.name):
        if paragraph.text.strip():
            text += paragraph.text + "\n"

    return text
```

```python
from tqdm import tqdm
from pptx import Presentation

def load_pptx(file_path):
    presentation = Presentation(file_path)

    text = ""

    for slide in tqdm(presentation.slides, desc=file_path.name):
        for shape in slide.shapes:
            if hasattr(shape, "text"):
                if shape.text.strip():
                    text += shape.text + "\n"

    return text
```

```python
from tqdm import tqdm

def load_txt(file_path):
    text = ""

    with open(file_path, "r", encoding="utf-8") as file:
        lines = file.readlines()

    for line in tqdm(lines, desc=file_path.name):
        text += line

    return text
```

```python
for file_path in tqdm(Path("../data/pdf").glob("*.pdf"), desc="PDFs"):
    documents.append(
        {
            "source": file_path.name,
            "type": "pdf",
            "content": load_pdf(file_path)
        }
    )

for file_path in tqdm(Path("../data/docx").glob("*.docx"), desc="DOCXs"):
    documents.append(
        {
            "source": file_path.name,
            "type": "docx",
            "content": load_docx(file_path)
        }
    )

for file_path in tqdm(Path("../data/pptx").glob("*.pptx"), desc="PPTXs"):
    documents.append(
        {
            "source": file_path.name,
            "type": "pptx",
            "content": load_pptx(file_path)
        }
    )

for file_path in tqdm(Path("../data/txt").glob("*.txt"), desc="TXTs"):
    documents.append(
        {
            "source": file_path.name,
            "type": "txt",
            "content": load_txt(file_path)
        }
    )
```
